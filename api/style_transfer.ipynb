```python
def plot_image(images, names=None, denorm=False):
    plt.figure(figsize=(8 * len(images), 10))

    for i in range(len(images)):
        if denorm:
            image = deNorm(images[i])
        else:
            image = images[i]

        plt.subplot(1, len(images), i + 1)

        if names == None:
            plt.title('Image {}'.format(i))
        else:
            plt.title(names[i])

        plt.imshow(image)
        plt.axis('off')

    plt.show()
```
